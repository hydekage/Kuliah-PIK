## Fungsi Kuadrat ##
$$f_x=ax^2+bx+c$$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
a,b,c,x = sp.symbols("a,b,c,x")
```

```python
display(a,b,c,x)
```

```python
fx = -a*x**2-b*x-c
f_diff_x = fx.diff(x)
display(fx)
display(f_diff_x)
```

```python
fn = sp.lambdify([a,b,c,x],fx)
```

```python
x_values = np.linspace(-10,10,41)
print(x_values)
```

```python
y_values = []
for i in x_values:
    y_values.append(fn(2,4,6,i))
print(y_values)
plt.plot(x_values,y_values)
plt.show()
```

```python
f,t = sp.symbols('f,t')
f_sin_t = a*sp.sin(2*3.14*f*t)
display(f_sin_t)
```

```python
f_sin_t_lambdify = sp.lambdify([a,f,t],f_sin_t)
```

```python
t_values = np.linspace(0,1,200)
```

```python
y_values = []
for i in t_values:
  y_values.append(f_sin_t_lambdify(2,100,i))
plt.plot(t_values,y_values)
plt.show()
```
