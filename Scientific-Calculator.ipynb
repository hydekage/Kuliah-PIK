```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
a,b,c,x = sp.symbols("a,b,c,x")
```

```python
display(a,b,c,x)
```

```python
fx = -a*x**2-b*x-c
fx2 = a*x+b
display(fx,fx2)
```

```python
fn = sp.lambdify([a,b,c,x],fx)
fn2 = sp.lambdify([a,b,x],fx2)
```

```python
x_values = np.linspace(-100,100,200)
```

```python
y_values = []
y_values2= []
for i in x_values:
    y_values.append(fn(1,2,3,i))
    y_values2.append(fn2(1,2,i))
plt.plot(x_values,y_values)
plt.show()
plt.plot(x_values,y_values2)
plt.show()
```
